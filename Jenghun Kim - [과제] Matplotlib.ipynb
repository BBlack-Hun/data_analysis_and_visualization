```python
# 패키지 로드
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 구글 드라이브와 연동이 disconnect 되었다면 아래 코드를 다시 실행해 줍니다.
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# 데이터 읽어 오기. pandas의 read_csv 사용
fname ='gdrive/My Drive/Colab Notebooks/data/tips.csv'
data = pd.read_csv(fname)
data
```

## 1. total_bill 과 tip 의 분포도(scatter) 시각화하기

### tip 금액에 따라 점의 크기 변화 주기, alpha 옵션으로 투명도 지정하기
### 축 label 추가하기, 플롯 title 추가하기


```python
plt.scatter(data.total_bill, data.tip, alpha=0.5)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('total_bill& tip\'s scatter')
```

## 2. 성별에 따른 tip 비교하기 (bar plat)

```python
mt = []
fmt = []
for sextip in zip(data.sex, data.tip):
  if sextip[0] == 'Female':
    fmt = np.append(fmt, sextip[1])
  else:
    mt = np.append(mt, sextip[1])

fig = plt.figure()
fig.subplots_adjust(hspace=0.5, wspace=0.5)


y_pos_m = np.arange(len(mt))
y_pos_f = np.arange(len(fmt))

ax1=fig.add_subplot(211)
ax1.bar(y_pos_m, mt, color = 'red', width=1.5, label= 'Man.tip')
ax1.xaxis.set_major_formatter(plt.NullFormatter())
plt.xlabel('Tip', c = 'black', fontweight='bold')
plt.ylim(0,10)


ax2 =fig.add_subplot(212)
ax2.bar(y_pos_f, fmt, width=1.5, label= 'FeMan.tip')
ax2.xaxis.set_major_formatter(plt.NullFormatter())
plt.xlabel('Tip', c = 'black', fontweight='bold')
plt.ylabel('Female', c = 'white', fontweight='bold')
plt.ylim(0,10);
```

## 3. 남자의 여자의 팁 금액의 분포도(histogram)를 그려 비교하세요

### 2 X 1 의 subplot을 이용하세요.

```python
fig = plt.figure()

ax = fig.add_subplot(2,1,1)
ax.hist(fmt, bins=20, color='y', alpha=0.3)

ax = fig.add_subplot(2,1,2)
ax.hist(mt, bins=20, color='k', alpha=0.3)
```
